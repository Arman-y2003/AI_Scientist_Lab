```python
import numpy as np
data = np.random.rand(10)
print("نمونه داده:", data)
print("میانگین:", np.mean(data))
print("انحراف معیار:", np.std(data))
```

```python
with open("results.txt", "w", encoding="utf-8") as f:
    f.write("نمونه داده:\n")
    f.write(str(data) + "\n")
    f.write("میانگین: " + str(np.mean(data)) + "\n")
    f.write("انحراف معیار: " + str(np.std(data)) + "\n")
```
